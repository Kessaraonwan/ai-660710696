```python
import pandas as pd
from datetime import datetime, timedelta
import numpy as np
```

```python
df = pd.read_csv('/student.csv')
df
```

```python
df = pd.DataFrame(df)
```

```python
# เพิ่มวันที่ลงทะเบียน (สำหรับ feature engineering)
base_date = datetime(2024, 1, 15)
df['วันที่ลงทะเบียน'] = [base_date + timedelta(days=np.random.randint(-30, 30))for _ in range(20)]
```

```python
print("="*50)
print("🎯 ข้อมูลหลัก (Master Dataset) ที่จะใช้ตลอดบทเรียน")
print("="*50)
print(df)
print("\n📊 สรุปปัญหาที่พบในข้อมูล:")
print(f"- ค่าหายทั้งหมด: {df.isnull().sum().sum()} ค่า")
print(f"- อายุผิดปกติ: {df[df['อายุ'] > 100]['อายุ'].values if any(df['อายุ'] > 100) else 'ไม่มี'}")
print(f"- GPA ผิดปกติ: {df[(df['GPA'] < 0) | (df['GPA'] > 4)]['GPA'].values if any((df['GPA'] < 0) | (df['GPA'] > 4)) else 'ไม่มี'}")
print(f"- รูปแบบการเขียนคณะ: {df['คณะ'].nunique()} แบบ")
print(f"- รูปแบบการเขียนเพศ: {df['เพศ'].nunique()} แบบ")
```

```python
# เก็บสำเนาข้อมูลต้นฉบับ
df_original = df.copy()  # สำคัญ! ใช้ .copy() เพื่อไม่ให้กระทบข้อมูลต้นฉบับ
```

```python
# วิเคราะห์ข้อมูล df ที่สร้างไว้
print("="*50)
print("📋 ข้อมูลทั่วไปของ Dataset")
print("="*50)
df.info()

print("\n" + "="*50)
print("📊 สถิติเชิงพรรณนา")
print("="*50)
print(df.describe())

# ตรวจสอบค่าหายในแต่ละคอลัมน์
missing_counts = df.isnull().sum()
print("\n❌ ค่าหายในแต่ละคอลัมน์:")
for col, count in missing_counts[missing_counts > 0].items():
    percentage = (count / len(df)) * 100
    print(f"   - {col}: {count} ค่า ({percentage:.1f}%)")
```

```python
# แก้ค่าผิดปกติของตัวเลขก่อน (เช่น อายุ > 100, GPA <0 หรือ >4)
df.loc[df['อายุ'] > 100, 'อายุ'] = np.nan
df.loc[(df['GPA'] < 0) | (df['GPA'] > 4), 'GPA'] = np.nan
df
```

```python
# เติม missing ข้อมูลตัวเลขด้วย median
num_cols = ['อายุ', 'GPA', 'ส่วนสูง_cm', 'น้ำหนัก_kg', 'ชั่วโมงเรียน_ต่อสัปดาห์']

for col in num_cols:
    print(f" คอลัมน์: {col}")

# หาแถวที่มี NaN ในคอลัมน์นี้
    missing_rows = df[df[col].isna()]

    print("ก่อนเติมค่า :")
    print(missing_rows[[col]])

    median_val = df[col].median()
    df[col].fillna(median_val, inplace=True)

 # แสดงค่าในแถวที่เคย NaN หลังเติม
    print(f"ค่า median ที่ใช้เติม: {median_val:.2f}")
    print("หลังเติมค่า :")
    print( print(df.loc[missing_rows.index, [col]]))
    print("\n")
```

```python
# เติมค่า missing ด้วย mode (ค่าที่เจอบ่อยที่สุด)
cat_cols = ['เพศ', 'คณะ']

for col in cat_cols:
    print(f" คอลัมน์: {col}")

    # แสดงก่อนเติมค่า (เฉพาะแถวที่ NaN)
    nan_rows = df[df[col].isna()]
    print("ก่อนเติมค่า:")
    print(nan_rows[[col]])

    # เติมค่าด้วย mode
    mode_val = df[col].mode()[0]
    df[col].fillna(mode_val, inplace=True)
    print(f"เติมค่า = {mode_val}")

    # แสดงหลังเติม (เฉพาะแถวเดิมที่เคย NaN)
    print("หลังเติมค่า:")
    print(df.loc[nan_rows.index, [col]])
    print("\n")
```

```python
# ลบแถวที่ชื่อเป็น NaN เพราะชื่อไม่ควรหาย
df.dropna(subset=['ชื่อ'], inplace=True)
df.reset_index(drop=True, inplace=True)
print("หลังลบ (เหลือ 18 คน):")
```

```python
# ทำให้ชื่อคณะเป็นมาตรฐานเดียวกัน
def standardize_faculty(name):
    if name in ['วิทยาศาสตร์', 'วิทย์', 'Science', 'คณะวิทยาศาสตร์','Sci']:
        return 'วิทยาศาสตร์'
    elif name in ['ศิลปศาสตร์', 'ศิลป์', 'Arts','Liberal Arts']:
        return 'ศิลปศาสตร์'
    else:
        return 'วิศวกรรมศาสตร์'
df['คณะ'] = df['คณะ'].apply(standardize_faculty)
```

```python
#  ทำให้เพศเป็นมาตรฐานเดียวกัน
def standardize_gender(gender):
    if gender in ['ชาย', 'ช', 'M', 'Male']:
        return 'ชาย'
    elif gender in ['หญิง', 'ญ', 'F', 'Female']:
        return 'หญิง'
    else:
        return 'อื่นๆ'

df['เพศ'] = df['เพศ'].apply(standardize_gender)
```

```python
# ทำ One-Hot Encoding และกำหนด dtype เป็น int
df_encoded = pd.get_dummies(df, columns=['คณะ'], dtype=int)

# แสดงผลลัพธ์
print(df_encoded.head())
```

```python
# สร้าง Features ใหม่จากวันที่ลงทะเบียน
df['ปีที่ลงทะเบียน'] = df['วันที่ลงทะเบียน'].dt.year
df['เดือนที่ลงทะเบียน'] = df['วันที่ลงทะเบียน'].dt.month
df['วันในสัปดาห์ที่ลงทะเบียน'] = df['วันที่ลงทะเบียน'].dt.day_name()
df['วันในสัปดาห์ที่ลงทะเบียน_ตัวเลข'] = df['วันที่ลงทะเบียน'].dt.weekday
df['ช่วงเวลาลงทะเบียนในเดือน'] = pd.cut(df['วันที่ลงทะเบียน'].dt.day,
                                         bins=[0, 10, 20, 31],
                                         labels=['ต้นเดือน', 'กลางเดือน', 'ปลายเดือน'])

# แสดงผลลัพธ์
print(df[['วันที่ลงทะเบียน',
          'ปีที่ลงทะเบียน',
          'เดือนที่ลงทะเบียน',
          'วันในสัปดาห์ที่ลงทะเบียน_ตัวเลข',
          'ช่วงเวลาลงทะเบียนในเดือน']].head())
```

```python
#การรวม Features
#สร้าง BMI และ สถานะสุขภาพ
df['BMI'] = df['น้ำหนัก_kg'] / (df['ส่วนสูง_cm'] / 100)**2
df['สถานะสุขภาพ'] = pd.cut(df['BMI'],
                            bins=[0, 18.5, 25, 30, 100],
                            labels=['ผอม', 'ปกติ', 'อ้วน', 'อ้วนมาก'])

# แสดงผลลัพธ์
print(df[['ชื่อ', 'ส่วนสูง_cm', 'น้ำหนัก_kg', 'BMI', 'สถานะสุขภาพ']].head())
```
