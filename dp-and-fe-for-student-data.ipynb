```python
# ‡πÄ‡∏Å‡πá‡∏ö‡∏™‡∏≥‡πÄ‡∏ô‡∏≤‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ï‡πâ‡∏ô‡∏â‡∏ö‡∏±‡∏ö
df_original = df.copy()  # ‡∏™‡∏≥‡∏Ñ‡∏±‡∏ç! ‡πÉ‡∏ä‡πâ .copy() ‡πÄ‡∏û‡∏∑‡πà‡∏≠‡πÑ‡∏°‡πà‡πÉ‡∏´‡πâ‡∏Å‡∏£‡∏∞‡∏ó‡∏ö‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ï‡πâ‡∏ô‡∏â‡∏ö‡∏±‡∏ö
```

```python
# ‡∏ß‡∏¥‡πÄ‡∏Ñ‡∏£‡∏≤‡∏∞‡∏´‡πå‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏• df ‡∏ó‡∏µ‡πà‡∏™‡∏£‡πâ‡∏≤‡∏á‡πÑ‡∏ß‡πâ
print("="*50)
print("üìã ‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ó‡∏±‡πà‡∏ß‡πÑ‡∏õ‡∏Ç‡∏≠‡∏á Dataset")
print("="*50)
df.info()

print("\n" + "="*50)
print("üìä ‡∏™‡∏ñ‡∏¥‡∏ï‡∏¥‡πÄ‡∏ä‡∏¥‡∏á‡∏û‡∏£‡∏£‡∏ì‡∏ô‡∏≤")
print("="*50)
print(df.describe())

# ‡∏ï‡∏£‡∏ß‡∏à‡∏™‡∏≠‡∏ö‡∏Ñ‡πà‡∏≤‡∏´‡∏≤‡∏¢‡πÉ‡∏ô‡πÅ‡∏ï‡πà‡∏•‡∏∞‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå
missing_counts = df.isnull().sum()
print("\n‚ùå ‡∏Ñ‡πà‡∏≤‡∏´‡∏≤‡∏¢‡πÉ‡∏ô‡πÅ‡∏ï‡πà‡∏•‡∏∞‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå:")
for col, count in missing_counts[missing_counts > 0].items():
    percentage = (count / len(df)) * 100
    print(f"   - {col}: {count} ‡∏Ñ‡πà‡∏≤ ({percentage:.1f}%)")
```

```python
# ‡πÅ‡∏Å‡πâ‡∏Ñ‡πà‡∏≤‡∏ú‡∏¥‡∏î‡∏õ‡∏Å‡∏ï‡∏¥‡∏Ç‡∏≠‡∏á‡∏ï‡∏±‡∏ß‡πÄ‡∏•‡∏Ç‡∏Å‡πà‡∏≠‡∏ô (‡πÄ‡∏ä‡πà‡∏ô ‡∏≠‡∏≤‡∏¢‡∏∏ > 100, GPA <0 ‡∏´‡∏£‡∏∑‡∏≠ >4)
df.loc[df['‡∏≠‡∏≤‡∏¢‡∏∏'] > 100, '‡∏≠‡∏≤‡∏¢‡∏∏'] = np.nan
df.loc[(df['GPA'] < 0) | (df['GPA'] > 4), 'GPA'] = np.nan
df
```

```python
# ‡πÄ‡∏ï‡∏¥‡∏° missing ‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ï‡∏±‡∏ß‡πÄ‡∏•‡∏Ç‡∏î‡πâ‡∏ß‡∏¢ median
num_cols = ['‡∏≠‡∏≤‡∏¢‡∏∏', 'GPA', '‡∏™‡πà‡∏ß‡∏ô‡∏™‡∏π‡∏á_cm', '‡∏ô‡πâ‡∏≥‡∏´‡∏ô‡∏±‡∏Å_kg', '‡∏ä‡∏±‡πà‡∏ß‡πÇ‡∏°‡∏á‡πÄ‡∏£‡∏µ‡∏¢‡∏ô_‡∏ï‡πà‡∏≠‡∏™‡∏±‡∏õ‡∏î‡∏≤‡∏´‡πå']

for col in num_cols:
    print(f" ‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå: {col}")

# ‡∏´‡∏≤‡πÅ‡∏ñ‡∏ß‡∏ó‡∏µ‡πà‡∏°‡∏µ NaN ‡πÉ‡∏ô‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå‡∏ô‡∏µ‡πâ
    missing_rows = df[df[col].isna()]

    print("‡∏Å‡πà‡∏≠‡∏ô‡πÄ‡∏ï‡∏¥‡∏°‡∏Ñ‡πà‡∏≤ :")
    print(missing_rows[[col]])

    median_val = df[col].median()
    df[col].fillna(median_val, inplace=True)

 # ‡πÅ‡∏™‡∏î‡∏á‡∏Ñ‡πà‡∏≤‡πÉ‡∏ô‡πÅ‡∏ñ‡∏ß‡∏ó‡∏µ‡πà‡πÄ‡∏Ñ‡∏¢ NaN ‡∏´‡∏•‡∏±‡∏á‡πÄ‡∏ï‡∏¥‡∏°
    print(f"‡∏Ñ‡πà‡∏≤ median ‡∏ó‡∏µ‡πà‡πÉ‡∏ä‡πâ‡πÄ‡∏ï‡∏¥‡∏°: {median_val:.2f}")
    print("‡∏´‡∏•‡∏±‡∏á‡πÄ‡∏ï‡∏¥‡∏°‡∏Ñ‡πà‡∏≤ :")
    print( print(df.loc[missing_rows.index, [col]]))
    print("\n")
```

```python
# ‡πÄ‡∏ï‡∏¥‡∏°‡∏Ñ‡πà‡∏≤ missing ‡∏î‡πâ‡∏ß‡∏¢ mode (‡∏Ñ‡πà‡∏≤‡∏ó‡∏µ‡πà‡πÄ‡∏à‡∏≠‡∏ö‡πà‡∏≠‡∏¢‡∏ó‡∏µ‡πà‡∏™‡∏∏‡∏î)
cat_cols = ['‡πÄ‡∏û‡∏®', '‡∏Ñ‡∏ì‡∏∞']

for col in cat_cols:
    print(f" ‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå: {col}")

    # ‡πÅ‡∏™‡∏î‡∏á‡∏Å‡πà‡∏≠‡∏ô‡πÄ‡∏ï‡∏¥‡∏°‡∏Ñ‡πà‡∏≤ (‡πÄ‡∏â‡∏û‡∏≤‡∏∞‡πÅ‡∏ñ‡∏ß‡∏ó‡∏µ‡πà NaN)
    nan_rows = df[df[col].isna()]
    print("‡∏Å‡πà‡∏≠‡∏ô‡πÄ‡∏ï‡∏¥‡∏°‡∏Ñ‡πà‡∏≤:")
    print(nan_rows[[col]])

    # ‡πÄ‡∏ï‡∏¥‡∏°‡∏Ñ‡πà‡∏≤‡∏î‡πâ‡∏ß‡∏¢ mode
    mode_val = df[col].mode()[0]
    df[col].fillna(mode_val, inplace=True)
    print(f"‡πÄ‡∏ï‡∏¥‡∏°‡∏Ñ‡πà‡∏≤ = {mode_val}")

    # ‡πÅ‡∏™‡∏î‡∏á‡∏´‡∏•‡∏±‡∏á‡πÄ‡∏ï‡∏¥‡∏° (‡πÄ‡∏â‡∏û‡∏≤‡∏∞‡πÅ‡∏ñ‡∏ß‡πÄ‡∏î‡∏¥‡∏°‡∏ó‡∏µ‡πà‡πÄ‡∏Ñ‡∏¢ NaN)
    print("‡∏´‡∏•‡∏±‡∏á‡πÄ‡∏ï‡∏¥‡∏°‡∏Ñ‡πà‡∏≤:")
    print(df.loc[nan_rows.index, [col]])
    print("\n")
```

```python
# ‡∏•‡∏ö‡πÅ‡∏ñ‡∏ß‡∏ó‡∏µ‡πà‡∏ä‡∏∑‡πà‡∏≠‡πÄ‡∏õ‡πá‡∏ô NaN ‡πÄ‡∏û‡∏£‡∏≤‡∏∞‡∏ä‡∏∑‡πà‡∏≠‡πÑ‡∏°‡πà‡∏Ñ‡∏ß‡∏£‡∏´‡∏≤‡∏¢
df.dropna(subset=['‡∏ä‡∏∑‡πà‡∏≠'], inplace=True)
df.reset_index(drop=True, inplace=True)
print("‡∏´‡∏•‡∏±‡∏á‡∏•‡∏ö (‡πÄ‡∏´‡∏•‡∏∑‡∏≠ 18 ‡∏Ñ‡∏ô):")
```

```python
# ‡∏ó‡∏≥‡πÉ‡∏´‡πâ‡∏ä‡∏∑‡πà‡∏≠‡∏Ñ‡∏ì‡∏∞‡πÄ‡∏õ‡πá‡∏ô‡∏°‡∏≤‡∏ï‡∏£‡∏ê‡∏≤‡∏ô‡πÄ‡∏î‡∏µ‡∏¢‡∏ß‡∏Å‡∏±‡∏ô
def standardize_faculty(name):
    if name in ['‡∏ß‡∏¥‡∏ó‡∏¢‡∏≤‡∏®‡∏≤‡∏™‡∏ï‡∏£‡πå', '‡∏ß‡∏¥‡∏ó‡∏¢‡πå', 'Science', '‡∏Ñ‡∏ì‡∏∞‡∏ß‡∏¥‡∏ó‡∏¢‡∏≤‡∏®‡∏≤‡∏™‡∏ï‡∏£‡πå','Sci']:
        return '‡∏ß‡∏¥‡∏ó‡∏¢‡∏≤‡∏®‡∏≤‡∏™‡∏ï‡∏£‡πå'
    elif name in ['‡∏®‡∏¥‡∏•‡∏õ‡∏®‡∏≤‡∏™‡∏ï‡∏£‡πå', '‡∏®‡∏¥‡∏•‡∏õ‡πå', 'Arts','Liberal Arts']:
        return '‡∏®‡∏¥‡∏•‡∏õ‡∏®‡∏≤‡∏™‡∏ï‡∏£‡πå'
    else:
        return '‡∏ß‡∏¥‡∏®‡∏ß‡∏Å‡∏£‡∏£‡∏°‡∏®‡∏≤‡∏™‡∏ï‡∏£‡πå'
df['‡∏Ñ‡∏ì‡∏∞'] = df['‡∏Ñ‡∏ì‡∏∞'].apply(standardize_faculty)
```

```python
#  ‡∏ó‡∏≥‡πÉ‡∏´‡πâ‡πÄ‡∏û‡∏®‡πÄ‡∏õ‡πá‡∏ô‡∏°‡∏≤‡∏ï‡∏£‡∏ê‡∏≤‡∏ô‡πÄ‡∏î‡∏µ‡∏¢‡∏ß‡∏Å‡∏±‡∏ô
def standardize_gender(gender):
    if gender in ['‡∏ä‡∏≤‡∏¢', '‡∏ä', 'M', 'Male']:
        return '‡∏ä‡∏≤‡∏¢'
    elif gender in ['‡∏´‡∏ç‡∏¥‡∏á', '‡∏ç', 'F', 'Female']:
        return '‡∏´‡∏ç‡∏¥‡∏á'
    else:
        return '‡∏≠‡∏∑‡πà‡∏ô‡πÜ'

df['‡πÄ‡∏û‡∏®'] = df['‡πÄ‡∏û‡∏®'].apply(standardize_gender)
```

```python
# ‡∏™‡∏£‡πâ‡∏≤‡∏á Features ‡πÉ‡∏´‡∏°‡πà‡∏à‡∏≤‡∏Å‡∏ß‡∏±‡∏ô‡∏ó‡∏µ‡πà‡∏•‡∏á‡∏ó‡∏∞‡πÄ‡∏ö‡∏µ‡∏¢‡∏ô
df['‡∏õ‡∏µ‡∏ó‡∏µ‡πà‡∏•‡∏á‡∏ó‡∏∞‡πÄ‡∏ö‡∏µ‡∏¢‡∏ô'] = df['‡∏ß‡∏±‡∏ô‡∏ó‡∏µ‡πà‡∏•‡∏á‡∏ó‡∏∞‡πÄ‡∏ö‡∏µ‡∏¢‡∏ô'].dt.year
df['‡πÄ‡∏î‡∏∑‡∏≠‡∏ô‡∏ó‡∏µ‡πà‡∏•‡∏á‡∏ó‡∏∞‡πÄ‡∏ö‡∏µ‡∏¢‡∏ô'] = df['‡∏ß‡∏±‡∏ô‡∏ó‡∏µ‡πà‡∏•‡∏á‡∏ó‡∏∞‡πÄ‡∏ö‡∏µ‡∏¢‡∏ô'].dt.month
df['‡∏ß‡∏±‡∏ô‡πÉ‡∏ô‡∏™‡∏±‡∏õ‡∏î‡∏≤‡∏´‡πå‡∏ó‡∏µ‡πà‡∏•‡∏á‡∏ó‡∏∞‡πÄ‡∏ö‡∏µ‡∏¢‡∏ô'] = df['‡∏ß‡∏±‡∏ô‡∏ó‡∏µ‡πà‡∏•‡∏á‡∏ó‡∏∞‡πÄ‡∏ö‡∏µ‡∏¢‡∏ô'].dt.day_name()
df['‡∏ß‡∏±‡∏ô‡πÉ‡∏ô‡∏™‡∏±‡∏õ‡∏î‡∏≤‡∏´‡πå‡∏ó‡∏µ‡πà‡∏•‡∏á‡∏ó‡∏∞‡πÄ‡∏ö‡∏µ‡∏¢‡∏ô_‡∏ï‡∏±‡∏ß‡πÄ‡∏•‡∏Ç'] = df['‡∏ß‡∏±‡∏ô‡∏ó‡∏µ‡πà‡∏•‡∏á‡∏ó‡∏∞‡πÄ‡∏ö‡∏µ‡∏¢‡∏ô'].dt.weekday
df['‡∏ä‡πà‡∏ß‡∏á‡πÄ‡∏ß‡∏•‡∏≤‡∏•‡∏á‡∏ó‡∏∞‡πÄ‡∏ö‡∏µ‡∏¢‡∏ô‡πÉ‡∏ô‡πÄ‡∏î‡∏∑‡∏≠‡∏ô'] = pd.cut(df['‡∏ß‡∏±‡∏ô‡∏ó‡∏µ‡πà‡∏•‡∏á‡∏ó‡∏∞‡πÄ‡∏ö‡∏µ‡∏¢‡∏ô'].dt.day,
                                         bins=[0, 10, 20, 31],
                                         labels=['‡∏ï‡πâ‡∏ô‡πÄ‡∏î‡∏∑‡∏≠‡∏ô', '‡∏Å‡∏•‡∏≤‡∏á‡πÄ‡∏î‡∏∑‡∏≠‡∏ô', '‡∏õ‡∏•‡∏≤‡∏¢‡πÄ‡∏î‡∏∑‡∏≠‡∏ô'])

# ‡πÅ‡∏™‡∏î‡∏á‡∏ú‡∏•‡∏•‡∏±‡∏û‡∏ò‡πå
print(df[['‡∏ß‡∏±‡∏ô‡∏ó‡∏µ‡πà‡∏•‡∏á‡∏ó‡∏∞‡πÄ‡∏ö‡∏µ‡∏¢‡∏ô',
          '‡∏õ‡∏µ‡∏ó‡∏µ‡πà‡∏•‡∏á‡∏ó‡∏∞‡πÄ‡∏ö‡∏µ‡∏¢‡∏ô',
          '‡πÄ‡∏î‡∏∑‡∏≠‡∏ô‡∏ó‡∏µ‡πà‡∏•‡∏á‡∏ó‡∏∞‡πÄ‡∏ö‡∏µ‡∏¢‡∏ô',
          '‡∏ß‡∏±‡∏ô‡πÉ‡∏ô‡∏™‡∏±‡∏õ‡∏î‡∏≤‡∏´‡πå‡∏ó‡∏µ‡πà‡∏•‡∏á‡∏ó‡∏∞‡πÄ‡∏ö‡∏µ‡∏¢‡∏ô_‡∏ï‡∏±‡∏ß‡πÄ‡∏•‡∏Ç',
          '‡∏ä‡πà‡∏ß‡∏á‡πÄ‡∏ß‡∏•‡∏≤‡∏•‡∏á‡∏ó‡∏∞‡πÄ‡∏ö‡∏µ‡∏¢‡∏ô‡πÉ‡∏ô‡πÄ‡∏î‡∏∑‡∏≠‡∏ô']].head())
```

```python
#‡∏Å‡∏≤‡∏£‡∏£‡∏ß‡∏° Features
#‡∏™‡∏£‡πâ‡∏≤‡∏á BMI ‡πÅ‡∏•‡∏∞ ‡∏™‡∏ñ‡∏≤‡∏ô‡∏∞‡∏™‡∏∏‡∏Ç‡∏†‡∏≤‡∏û
df['BMI'] = df['‡∏ô‡πâ‡∏≥‡∏´‡∏ô‡∏±‡∏Å_kg'] / (df['‡∏™‡πà‡∏ß‡∏ô‡∏™‡∏π‡∏á_cm'] / 100)**2
df['‡∏™‡∏ñ‡∏≤‡∏ô‡∏∞‡∏™‡∏∏‡∏Ç‡∏†‡∏≤‡∏û'] = pd.cut(df['BMI'],
                            bins=[0, 18.5, 25, 30, 100],
                            labels=['‡∏ú‡∏≠‡∏°', '‡∏õ‡∏Å‡∏ï‡∏¥', '‡∏≠‡πâ‡∏ß‡∏ô', '‡∏≠‡πâ‡∏ß‡∏ô‡∏°‡∏≤‡∏Å'])

# ‡πÅ‡∏™‡∏î‡∏á‡∏ú‡∏•‡∏•‡∏±‡∏û‡∏ò‡πå
print(df[['‡∏ä‡∏∑‡πà‡∏≠', '‡∏™‡πà‡∏ß‡∏ô‡∏™‡∏π‡∏á_cm', '‡∏ô‡πâ‡∏≥‡∏´‡∏ô‡∏±‡∏Å_kg', 'BMI', '‡∏™‡∏ñ‡∏≤‡∏ô‡∏∞‡∏™‡∏∏‡∏Ç‡∏†‡∏≤‡∏û']].head())
```
